Before we can begin cleaning our data, we'll need to import the pandas library.

```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

Now, let's create a DataFrame called "df_census" and load in our data. 

```python
file_census = "../data_archive/census_data_2022_03_01.tgz"
df_census = pd.read_csv(file_census, low_memory=False, compression='gzip')
```

Using .shape, we can tell before even laying eyes on the DataFrame that is going to be enormous. 73 columns and 85,000+ rows!

```python
df_census.shape
```

We can get a pretty good sense of what we're working with by viewing just the first few rows of our DataFrame. Som let's use .head(), to limit the display to just the first 5 rows.

```python
df_census.head()
```

The columns have been named using Census Bureau codes that aren't really meaningul to us. Fortunately, descriptions of each column's data have been provided in the first row (index 0). We'll use these to rename our columns, but first, let's use regex to remove all of those distracting exclamation points! 

```python
df_census.iloc[0].replace('!!', ' ', regex=True, inplace=True)
df_census.iloc[0].str.lstrip()
df_census.head()
```

Patrick - I can't figure out why the column names still have white spaces at the front, I used strip() on the row before turning into column headers. 

```python
df_census.columns = df_census.iloc[0]
df_census = df_census.drop(0)
df_census.rename(columns={
    'id': 'GEOID Full',
    'Geographic Area Name': 'Tract Name',
    '  Total:': 'Population'},
    inplace=True)
df_census.head()
```

Now that our headers give a better sense of the data contained in each column, we can see that we really only need that first column of population data. We can "throw away" the other 70 columns! So, let's create a new, much smaller DataFrame with just the columns we need. We'll call it "df_census_pop".

```python
df_census_pop = df_census[['GEOID Full', 'Tract Name', 'Population']]
df_census_pop.head()
```

Let's run .info() on our new DataFrame to see what we might need to cleanup.

```python
df_census_pop.info()
```

We've learned from .info that all of our columns are of datatype object. Based on our visual inspection of the first few rows of data, we know that the values in both columns GEOID Full and Tract Name are actually strings, while values in the column Population are really integers. We'll fix this in a moment. First, we want to deal with any null values. 

According to .info, all of the columns are free of null values. Again, we're a little skeptical of .info and since we need to be sure this is actually the case, so we'll probe further.

Okay, now let's load in the dataset containing the area in square miles of each census tract. We're going to match this up to our first dataframe, so we can calculate population density. 

From previous research, we know that the GEOID Structure for a census tract is: STATE+COUNTY+TRACT | Digits: 2+3+6=11 | Example GeoArea: Census Tract 2231 in Harris County, TX | Example GEOID: 48201223100

Having taken a peek at the data before loading it, we know that the GEOIDS here are for the census tract level. So, we're going to rename our column to indicate this. 

```python
file_gaz = "../data_archive/2020_Gaz_tracts_national.gz"
df_gaz = pd.read_csv(file_gaz, delimiter = '\t', compression='gzip')
df_gaz.rename(columns={'GEOID' : 'GEOID Tract'},
    inplace=True)
df_gaz.head(10)
```
