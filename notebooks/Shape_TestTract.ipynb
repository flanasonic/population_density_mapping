Federal Information Processing Standards (FIPS) code

Here, the "GEOID" is a census tract FIPS code, an 11-digit number that uniquely identifies each census tract in the United States. It is a concatenation of the 2-digit state code, 3-digit county code, and 6-digit tract code. So, we can break down a GEOID like this:
"GEOID" = "STATEFP" + "COUNTYFP" + "TRACTCE"

STATEFP = 2-digit state FIPS code
COUNTYFP = 3-digit county FIPS code  
TRACTCE = 6-digit census tract code

Example:
"GEOID": "50005957100" 
"STATEFP": "50" - Vermont
"COUNTYFP":  "005" - Caledonia
"TRACTCE": "957100" 

So, the above example is Census Tract 957100, which is in Caledonia County, in the state VT
